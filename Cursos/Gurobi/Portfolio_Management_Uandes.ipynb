# Portfolio Management

- Markewitz 1952 -> Portafolio eficiente
- ¿Que es el portafolio eficiente?
 - Es una collecion de diferentes activos e instrumentos financieros que tienen por objetivo sacar cierta rentabilidad de mercado
 - La creacion de un portafolio es una estrategia de diversificación donde lo qye se busca es **minimizar el riesgo que optimice los retornos.**

- ¿Cuando es eficiente?
 - Cuando para un rendimiento esperado, no existe ninguna cartera que tnega un riesgo mas bajo.
 - Para un determinado riesgo, no existe otra opcion que me entregue un retorno mayor.

## ¿Que es el rendimeinto?

- Porcentaje de ganacia o utilidad que se obtiene respecto a una inversion para un $t$
- En general el rendimiento es mayor a un mayor plazo.



## ¿Que es el riesgo?

- Es la incertidumbre producida en el rendimiento de la inversion

El modelo de Markewitz se basa en la racionalidad del inversionista.
 - Maximizar rendimiento al minimo riesgo.

La cartera optima esta en el punto de tangencia entre la F.E. (Frontera Eficiente) y la curva de indeferencia mas elevada.

Matematicamente: Riesgo = Varianza

## ¿Que es la varianza?

Grado de dsipersión de un conjunto de valores/observaciones (acciones) respecto a un valor central (respecto a su media)

$$
\sigma =\sqrt{var(R_{i})} = \sqrt{E(R_{i} - \bar{R_{i}})^2} = \sqrt{\frac{\sum_{t=1}^{T}(R_{i} - \bar{R_{i}})^2}{T}}
$$

Teniendo como datos los retornos historicos $R_{i}$ -> $R_{i}^t$

## ¿Que es la covarianza?


Es el grado en que dos variables (acciones) se mueven en la misma direccion o direcciones opuestas, una respecto a la otra.

$$
COV(X, Y) = E((X - E(X))(Y - E(Y))) = \rho_{(X, Y)} \sigma_{X}\sigma_{Y}
$$


## ¿Que es el coeficiente de correlacion ($\rho$)?

- Nos permite observar como se comporta una accion $i$ respect a una accion $j$. (El rendimiento de ellas)

|coef|comportamiento|
|----|--------------|
|+1|las 2 acciones tienen comportamiento simiar|
|0|son independientes|
|-1|tienen comportamiento inverso|

image.png

- Uno prefiere la **Correlacion (-)**, ya que si a alguna accion le empieza a ir mal, se compensa con otra accion a la que le estarà llendo mejor.

## ¿y como medimos la rentabilidad?


**Retorno esperado**

$$
\bar{R_{i}} = E(R_{i}) = \frac{1}{T} \sum_{t=1}^{T} R_{i}^{t}
$$

## Hipotesis modelo de markowitz

- El rendimiento de cualquier portafolio se considera una variable aleatoria. El valor que se espera a utilizar para cuantificar la rentabilidad del portafolio.

- La desviacion estandar y la varianza son tomados como medidas de dispersion, es decir, como medidas de riesgo de la variable aleatoria.

- La decision del inversionista esta por el camino de optar por un portafolio que le genere el mayor retorno asumiendo un minimo riesgo. 

- **Extension a N activos**

**Retorno Portafolio**

$$
R_{p} = \sum_{i=1}^{N} w_{i}E(R_{i})
$$

**Riesgo Portafolio**

$$
\sigma_{P} =\sqrt{\sum_{i=1}^{N}\sum_{j=1}^{N}w_{i}w_{j}\sigma_{i}\sigma_{j}\rho_{(i, j)}}
$$

## Formulacion de un portafolio optimo

Una de las formas de encontrar este conjunto de instrumentos financieros (acciones) de manera eficiente es por medio de un problema de optimizacion donde se considera una formulacion dual:

- 1) Determinar los pesos ($w_{i}$) **que minimizan la variannza** sujeto a un retorno requerido ($R_{P}^*$)

  - **F.O.**

    $$
    min \quad var(R_{p})=\sigma_{p}^{2} = \sum_{i=1}^{N}\sum_{j=1}^{N}w_{i}w_{j}\sigma_{i}\sigma_{j}\rho_{(i, j)}
    $$

  - **S.A.**

    $$
    \sum_{i=1}^{N} w_{i}\bar{R_{i}} \geq R_{p}^{*}
    $$

    $$
    \sum_{i=1}^{N} w_{i} = 1
    $$

    $$
    w_{i} \geq 0 \quad \forall i \in (1, ..., N)
    $$

  - Siendo $\bar{R_{i}}$ = Rendimiento esperado del activo $i$ 

- 2) Consiste en determinar los pesos ($w_{i}$) que maximizan el retorno esperado ($R_{P}$) sujeto a un maximo riesgo ($R_{p}^{*}$)

  - **F.O.**

  $$
  max \quad R_{P} = \sum_{i=1}^{N}w_{i}\bar{R_{i}}
  $$

  - **S.A.**

  $$
  \sum_{i=1}^{N}\sum_{j=1}^{N}w_{i}w_{j}\sigma_{i}\sigma_{j}\rho_{(i, j)} \leq \sigma_{p}^{2}
  $$

  $$
  \sum_{i=1}^{N}w_{i} = 1 
  $$

  $$
  w_{i} \geq 0 \quad \forall i \in (1, ..., N)
  $$



En cualquiera de los dos casos (optimizando la varianza o el retorno esperado), se encontraran los pesos de los activos ($w_{i}$) que optimizarán el portafolio sujeto a las restricciones dadas, generando así la curva eficiente de portafolio.

- Una vez resuelto el problema, obtengo la proporcion a invertir en cada activo.

- Algunas restricciones adicionales:
 - Restringir la inversion 
 - Agrupacion de acciones 
 - restringir la proporcion de una accion 
 - relajar la restriccion de positividad de los pesos 

**Construccion de un portafolio optimo**

- 1) Seleccion de N acciones y sus valores en el tiempo 
- 2) Calcular la rentabilidad ($R_{i}^{t}$) y su varianza ($\sigma_{i}^{2}$) para cada activo.
- 3) construyo un vector de pesos ($w_{i}$)
- 4) Calculo $R_{P}=\sum_{i=1}^{N}w_{i}\bar{R_{i}}$ y $\sigma_{P}^{2} = \sum_{i=1}^{N}\sum_{j=1}^{N} w_{i}w_{j}\sigma_{i}\sigma_{j}\rho_{(i, j)}$
- 5) Resolver el problema de optimización.

## Extract $\bar{R}$ and $\sigma$ (Yahoo Finance)

```python
import pandas as pd
import numpy as np
!pip install yfinance
import yfinance as yf
```

```python
# funcion para retornar la esperanza y la varianza de la accion
# Parametros:
# Nombre de la accion 
# Periodo de tiempo 

def R_and_Sigma(nombre_accion, period='5y'):

  ticker = yf.Ticker(nombre_accion)
  df = ticker.history(period=period)

  df = df['Close'].to_frame()
  df['date'] = df.index
  df['Month-Year'] = df['date'].apply(lambda x:pd.Timestamp(x).strftime('%Y-%m'))

  df = df.groupby(by=['Month-Year']).mean().reset_index() # Valor promedio del mes
  df['Month-Year'] = pd.to_datetime(df['Month-Year'], infer_datetime_format=True)
  df = df.set_index(['Month-Year'])

  df1 = df['Close'].copy()
  df_Rt = (df1/df1.shift())-1
  df_Rt = df_Rt.dropna()

  df_Rt = df_Rt.to_frame()

  df_Rt['dates'] = df_Rt.index
  E_Rt = df_Rt['Close'].mean()
  sigma = df_Rt['Close'].std()

  return np.round(E_Rt, 3), np.round(sigma, 3), np.array(df['Close'])
```

```python
E_Rt, sigma, prices = R_and_Sigma('MCD', period='5y')
print(f'E_Rt: {E_Rt}')
print(f'Sigma: {sigma}')
print(prices)
```

## Solve model

```python
!pip install gurobipy 
from gurobipy import Model, quicksum, GRB
```

```python
# Datos
portafolio = ['MCD', 'TSLA', 'PG', 'STLA', 'PEP', 'KO', 'WMT', 'MSFT', 
              'IBM', 'LTMAQ', 'GOOG', 'AMZN', 'FB', 'NVDA', 'V', 'JNJ', 
              'NSRGY', 'MA', 'TM', 'DIS', 'ADBE', 'NKE', 'INTC', 'PYPL']
dic_rent_sigma = {i:R_and_Sigma(i, period='3y')[:2] for i in portafolio}
min_var = (0.025)**2
dic_rent_sigma
```

```python
dic_prices = {i:R_and_Sigma(i, period='3y')[2] for i in portafolio}
```

```python
dic_corr = {(i, j):np.corrcoef(dic_prices[i], dic_prices[j])[0][1] for i in portafolio for j in portafolio}
for i in portafolio[:3]:
  for j in portafolio[:3]:
    print(f'corr {i} {j} = {dic_corr[i, j]}')
```

```python
# Modelo
m = Model()

# Variables 
w = {}
for i in portafolio:
  w[i] = m.addVar(lb=0.0, ub=1.0, vtype='C')

obj = m.addVar(vtype='C')

# F.O. 
m.addConstr(obj == quicksum(dic_rent_sigma[i][0]*w[i] for i in portafolio))

# Restricciones 
m.addConstr(quicksum(w[i]*w[j]*dic_rent_sigma[i][1]*dic_rent_sigma[j][1]*dic_corr[i, j] for i in portafolio for j in portafolio) <= min_var)

m.addConstr(quicksum(w[i] for i in portafolio) == 1.0)

m.setObjective(obj, GRB.MAXIMIZE)
m.optimize()
```

```python
for i in portafolio:
  print(f'{i}: {np.round(w[i].x*100, 2)}%')

print(f'\n Rent Esp: {obj.x}')
```

## Frontera Eficiente


```python
def portfolio_model(portafolio, min_var):
  # Modelo
  m = Model()

  # Variables 
  w = {}
  for i in portafolio:
    w[i] = m.addVar(lb=0.0, ub=1.0, vtype='C')

  obj = m.addVar(vtype='C')

  # F.O. 
  m.addConstr(obj == quicksum(dic_rent_sigma[i][0]*w[i] for i in portafolio))

  # Restricciones 
  m.addConstr(quicksum(w[i]*w[j]*dic_rent_sigma[i][1]*dic_rent_sigma[j][1]*dic_corr[i, j] for i in portafolio for j in portafolio) <= min_var)

  m.addConstr(quicksum(w[i] for i in portafolio) == 1.0)

  m.setObjective(obj, GRB.MAXIMIZE)
  m.optimize()

  return obj.x
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.01, 0.05, 50)
y = [portfolio_model(portafolio, i) for i in x]
```

```python
plt.plot(x, y)
plt.title('Frontera Eficiente')
plt.xlabel('Riesgo Portafolio')
plt.ylabel('Rendimiento Portafolio')
plt.show()
```
