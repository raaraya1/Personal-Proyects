# Resumen Series de Tiempo

-	En teoría de probabilidades, se define serie de tiempo como una **secuencia de variables aleatorias**.
-	En estadística, es una **variable que depende del tiempo (o que vaya evolucionando en el tiempo)**:
 -	El precio de una acción, que cambia cada día
 - Las pulsaciones de un paciente, medidas minuto a minuto
-	Se utilizan para predecir lo que ocurrirá con una variable en el futuro a partir del comportamiento de esa variable en el pasado
 -	Permite planificar/prevenir (planificar la producción, cuanto debo dejar en bodega)
 -	Ligado a RM/manejo de inventarios (estimación de demanda) (para manejar el inventario debo saber cuánto voy a vender)

Ejemplos: El precio del dólar

### Tendencia


-	La tendencia es un **movimiento de larga duración que muestra la evolución general** de la serie en el tiempo (valor promedio que se va moviendo a través del tiempo)

-	Puede ser estacionaria, creciente o decreciente

image.png

### Estado estacionario

-	Se dice que una serie de tiempo es estacionaria cuando el valor de su media, varianza y covarianza no varían sistemáticamente en el tiempo.

image.png

**Variaciones**

-	**Estacional**: movimiento periódico de corto periodo, se trata de una componente causal debida a la influencia de ciertos fenómenos que se repiten de manera periódica en un año (las estaciones), una semana (los fines de semana) o un día (las horas punta) o cualquier otro periodo

-	**Cíclica**: es el componente de la serie que recoge las oscilaciones periódicas de amplitud **superior a un año**. Movimientos normalmente irregulares alrededor de la tendencia, en las que, a diferencia de las variaciones estacionales, tiene un periodo y amplitud variables.

-	**Aleatoria**: de carácter errático, también denominada residuo, no muestra ninguna regularidad

## Suavizamiento exponencial

-	Se basa en la idea de que es posible calcular un promedio nuevo a partir de un promedio anterior y también del último dato observado

En el pronostico actual, le vamos a creer mas a la ultima demanda o a la historia, si alfa vale 1 le creo mas a la ultima demanda y si alfa es cercana a 0 le creo más a la historia.

**Este pronostico funciona bien para series de tiempo cuando no hay presencia de tendencia ni de estacionalidad**.

image.png

-	Alfa es la proporción del peso que se da a la última demanda contra la que se da al promedio anterior
-	Mayor alfa implica más nos acercamos al valor de la demanda que se acaba de observar … se les da mayor peso a las observaciones recientes que al promedio anterior.

**El método de Holt-Winters para pronósticos de corto plazo** es bastante bueno, ya que para hacer pronósticos a mas largo plazo se queda costo de información.

## Métodos Box Jenkins (1976)

-	Box y Jenkins desarrollaron modelos estadísticos que tienen en cuenta la dependencia existente entre los datos (los datos están relacionados con los datos anteriores)
-	Cada observación en un momento dado es modelada en función de los valores anteriores
-	Ignora la información de variables causales (por ejemplo, la venta de paraguas con la lluvia)
-	Procedimiento técnicamente sofisticado de predicción de una variable
-	Utiliza la observación más reciente como valor inicial
-	La autocorrelación es la correlación cruzada de una serie consigo misma
-	Analiza errores recientes de pronósticos para seleccionar el ajuste apropiado para periodos futuros (la parte MA(q), de medias móviles)
-	Box-Jenkins es más apropiado para predicciones a largo plazo que para costo plazo
-	Extrae mucha información de la serie de tiempo, más que cualquier otro método
-	Existen tres tipos básicos de modelos a ser examinados
-	Modelos autorregresivos (AR)
 -	Modelos de medias móviles (MA)
 -	Modelos mixtos autorregresivos-medias móviles (ARIMA)

-	Los modelos mixtos son una combinación lineal de los 2 anteriores y en estos la variable dependiente depende tanto de valores anteriores de la misma variable como de los valores anteriores de su error.

### Método combinado ARIMA(p, d, q)

-	Autoregressive integrated moving average ARIMA(p, d, q): se aplican diferencias para convertirlo en estacionario (generalización de ARMA):

image.png

El **“d” se aplica para dejar una serie como estacionaria**, se utiliza para eliminar la tendencia de la serie, para esto aplica un factor diferencial. Hay que determinar cuantos diferenciales se deben ocupar para dejar la serie en forma estacionaria.

-	Seasonal ARIMA -> SARIMA(p, d, q)(sp, sd, sq): Para la serie que presentan estacionalidad se pueden reproducir los mismos procesos sobre el orden estacional s(s=4 trimestrales, s=12 mensuales)


El **SARIMA le agrega la estacionalidad a la predicción.**

image.png

Por ejemplo, toma el sábado pasado, para el caso de el modelo autoregresivo, o toma el error del sábado pasado para el caso del modelo de medias móviles.

**Resumen parámetros (p, d, q) (sp, sd, sq) [m]**

-	P=cantidad de valores pasados que ira a buscar para la autorregresion

-	D=cantidad diferencias que son necesarias para convertir la serie original en estacionaria

-	Q=cantidad de errores pasados que ira a buscar par la media móvil

-	Sp=cantidad de periodos estacionales a considerar con autoregresion

-	Sd=cantidad de periodos estacionales a considerar en la cantidad de diferencias

-	Sq=cantidad de periodos estacionales a considerar con media móvil

-	M=periodo de la estacionalidad principal

-	SARIMAX: modelo SARIMA que además incluye variables exógenas, las cuales se incluyen con sus respectivos coeficientes
