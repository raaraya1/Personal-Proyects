17) Como puede impactar un ciclo de retroalimentacion en relacion a su politica predictiva?

20) Que es el "validation set"? que es el "test set"? porque son necesarios?

- "validation set": es el conjunto de datos que separamos de todos los disponibles y los utilizamos, unicamente, para medir el desempeño del modelo.

- "test set": Tambien correspoden a datos que separamos de todos los disponibles, y estos solo los llegamos a ocupar luego de que nos decidimos por un modelo.

El 'validation set' se utiliza para evitar un fenomeno llamado "over-fitting".
Este fenomeno se genera de realizar un entrenamiento excesivo con los datos del "training set", con lo cual el modelo ajusta perfectamente sus parametros para realizar prediccion certera sobre su conjunto de entrenamiento. Lo malo de esta historia, es que como estos parametros estan tan ajustados en el modelo, a este le resulta dificil adaptarse o predecir sobre algo que nunca a visto (como el "validation set").

Asimismo, el 'test set' se utiliza unicamente al final del trabajo y sucede que se hace esta separacion para tener una medicion mas justa sobre el desempeño del modelo elaborado. (en la busqueda de generar el mejor modelo ajustando los hyperparametros, podria originarse un modelo que este mas orientado a tener buenos resultados sobre los datos de validacion. Por eso se hace esta separacion, y asi tener una comparacion mas justa sobre el modelo con sus variantes, en sus hyperparametros)

23) Que es el 'overfitting'? Da un ejemplo.

Justamente como lo indica su nombre, es cuando un modelo se sobreajusta. Lo que sucede cuando entrenamos un modelo de redes neuronales, es que vamos ajustando sus parametros para ir haciendo cada vez menor la perdida (loss) sobre el conjunto de datos de entrenamiento. Lo malo es que si al modelo lo entrenamos mucho, puede que llegue a ser un experto sobre el conjunto de datos de entrenamiento, pero a la hora de utilizar un conjunto de nuevo (como el validation set) este no arrojara buenos resultados, ya que en su proceso de entrenamiento habra perdido su capacidad de generalizar conocimiento.

En cuanto al ejemplo, en la clase muestran como una curva cuadratica puede predecir mejor un comportamiento futuro, que si utilizamos un modelo que se ajuste a todos los puntos. Asimismo, y llevandolo a un problema mas cotidiano, yo lo pienso con el tamaño y forma de los zapatos, en el sentido de que si hago unos zapatos a la medida (overfitting) estos solo lo podra utilizar una unica persona. En cambio, si genero un par de zapatos mas olgados, estos lo podran utilizar mas personos, a pesar de que algunos puedan tropezar en ocaciones.

### Este es otro de los modelos que encontre en el notebook de la primera clase. Se trata de un localizador de objetos que utiliza como datos de entrada videos de calles y autopistas. De esta manera, se busca identificar objetos como personas y autos, lo cual luego podria contribuir en la elaboracion de un vehiculo que se maneje de forma autonoma.

````python
path = untar_data(URLs.CAMVID_TINY)
dls = SegmentationDataLoaders.from_label_func(
    path, bs=8, fnames = get_image_files(path/"images"),
    label_func = lambda o: path/'labels'/f'{o.stem}_P{o.suffix}',
    codes = np.loadtxt(path/'codes.txt', dtype=str)
)

learn = unet_learner(dls, resnet34)
learn.fine_tune(8)
````

### Estos son los resultados que arroja el modelo. Los cuadros a la derecha correponden a las imagenes coloreadas por el modelo.

````python
learn.show_results(max_n=6, figsize=(7,8))
````

# **Analizemos el tercer codigo de la clase1 (clasificador de sentimientos)**

````python
# Hagamos alguno ejemplo de los aparece en la pagina https://github.com/fastai/fastprogress
import numpy as np
mb = master_bar(range(10))
mb.names = ['cos', 'sin']
for i in mb:
    for j in progress_bar(range(100), parent=mb):
        if j%10 == 0:
            k = 100 * i + j
            x = np.arange(0, 2*k*np.pi/1000, 0.01)
            y1, y2 = np.cos(x), np.sin(x)
            graphs = [[x,y1], [x,y2]]
            x_bounds = [0, 2*np.pi]
            y_bounds = [-1,1]
            mb.update_graph(graphs, x_bounds, y_bounds)
            mb.child.comment = f'second bar stat'
    mb.main_bar.comment = f'first bar stat'
    mb.write(f'Finished loop {i}.')
````
