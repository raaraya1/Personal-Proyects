```python
y = {'f(x)':[2, 4, 6, 8], 'g(x)':[3, 5, 2, 8]}
x = [1, 2, 3, 4]

graph = pd.DataFrame(y, x)
graph.plot(kind='line', title='grafico de funciones', grid=True, xlim=(1, 4), xlabel='x', ylabel='y')
```

```python
# ahora, para guardar la imagen
grafico = graph.plot(kind='line', 
                     title='grafico de funciones', 
                     grid=True, 
                     xlim=(1, 4), 
                     xlabel='x', 
                     ylabel='y').get_figure()
# para guardar el grafico
ruta = os.getcwd()
grafico.savefig(ruta + '\grafico.jpg')
```
