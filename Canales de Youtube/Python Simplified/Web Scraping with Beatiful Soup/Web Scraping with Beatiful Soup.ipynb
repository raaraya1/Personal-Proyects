```python
# Bibliotecas a ocupar
import urllib.request
# para instalar pip install beautifulsoup4
from bs4 import BeautifulSoup as bs
import re
import pandas as pd
```

```python
# cargar un codigo html desde una url
page = urllib.request.urlopen('https://docs.python.org/3/library/random.html')
soup = bs(page)

# buscar el nombre de todas las funciones
names = soup.body.findAll('dt')
function_names = re.findall('id="random.\w+', str(names))
function_names = [item[4:] for item in function_names]

# buscar la descripcion de las funciones
description = soup.body.findAll('dd')
function_usage = []

for item in description:
    item = item.text
    item = item.replace('\n', ' ')
    function_usage.append(item)

print(function_names)
print(function_usage)
print(len(function_names))
print(len(function_usage))

```

```python
data = pd.DataFrame({'function names':function_names , 'function usage':function_usage})
data
```

```python
# buscar caracteristicas en especifico
example = soup.body.findAll('div', attrs={'id':'bookkeeping-functions'})
print(example)
```

```python
# luego para exportar como un archivo csv
data.to_csv('my_file.csv')
```
